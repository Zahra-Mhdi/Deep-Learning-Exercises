 Exercise 4-A : Implement Linear Regression from Scratch (PyTorch or NumPy)
- y = 3x + 2
- Compare learning rates [0.001, 0.01, 0.1]
- Use Adam optimizer

```python
# lr=0.001
import torch
from torch import nn
import matplotlib.pyplot as plt

torch.manual_seed(0)
x = torch.linspace(-3,3,100).unsqueeze(1)
y = 3*x + 2 + torch.randn(x.size())

model = nn.Linear(1,1)
opt = torch.optim.Adam(model.parameters(), lr=0.001)
loss_function = nn.MSELoss()
losses = []
for epoch in range(100):
  y_predicted = model(x)
  loss = loss_function(y_predicted, y)
  losses.append(loss.item())
  opt.zero_grad()
  loss.backward()
  opt.step()
print("Learned W and b for lr = 0.001", [p.data for p in model.parameters()])

plt.plot(losses)
plt.title("Training Loss (lr = 0.001)")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.show()
```

 Exercise 4-B : Fit nonlinear function y = 2x^2 + 1

```python
import torch
from torch import nn
import matplotlib.pyplot as plt

torch.manual_seed(0)
x = torch.linspace(-3,3,100).unsqueeze(1)
X = torch.cat([x, x**2], dim=1)
y2 = 2*x**2 + 1 + torch.randn(x.size())

model2 = nn.Linear(2,1)
opt2 = torch.optim.Adam(model2.parameters(), lr=0.01)
loss_function2 = nn.MSELoss()
losses2 = []
for epoch in range(100):
  y_predicted2 = model2(X)
  loss2 = loss_function2(y_predicted2, y2)
  losses2.append(loss2.item())
  opt2.zero_grad()
  loss2.backward()
  opt2.step()
print("Learned W and b", [p.data for p in model2.parameters()])

plt.plot(losses2)
plt.title("Training Loss (y = 2x^2 + 1)")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.show()
```
