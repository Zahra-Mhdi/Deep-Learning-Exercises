 Exercise 1-A : Compute y=Wx+b with MSE loss.

```python
import numpy as np
x=np.array([[1,2]])
print('x=',x)
w=np.array([[2],[3]])
print('w=',w)
b=np.array([1])
print('b=',b)
y=x@w+b
print('y=',y)
target=10
loss=np.mean((y-target)**2)
print('Loss=',loss)
```

 Exercise 1-B : Compute y=Wx^2+b with MSE loss.

```python
import numpy as np
x=np.array([[1,2]])
print('x=',x)
w=np.array([[2],[3]])
print('w=',w)
b=np.array([1])
print('b=',b)
x_square=x**2
print('x_square=',x_square)
y=x_square@w+b
print('y=',y)
target=10
loss=np.mean((y-target)**2)
print('loss=',loss)
```
